```python
# get active game ids
def get_active_game_ids():
    active_game_ids = []
    for i in range(len(today_schedule)):
        if today_schedule[i]['status'] == 'In Progress':
            active_game_ids.append(today_schedule[i]['game_id'])
    return active_game_ids
active_game_ids = get_active_game_ids()
active_game_ids
```
